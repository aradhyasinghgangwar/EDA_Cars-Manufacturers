```python
df = df[['Make','Model','Ex-Showroom_Price','Body_Type','Displacement', 'Cylinders', 'Fuel_System', 'Fuel_Tank_Capacity', 'Fuel_Type',  'City_Mileage',  'Seating_Capacity','Basic_Warranty', 'Child_Safety_Locks', 'Number_of_Airbags', 'Parking_Assistance',  'Navigation_System']]
```

```python
df.duplicated().any
```

```python
df['Ex-Showroom_Price'] = df['Ex-Showroom_Price'].str.replace(',','')
```

```python
df['Ex-Showroom_Price'] = df['Ex-Showroom_Price'].str.replace('Rs. ','')
```

```python
df['Displacement']=df['Displacement'].str.replace('cc','')
df['Displacement']
```

```python
df['City_Mileage'] = df['City_Mileage'].str.replace('?',' ')
```

```python
df['City_Mileage']=df['City_Mileage'].str.strip('km/litre')
df['City_Mileage']=df['City_Mileage'].str.replace(',','.')
df['City_Mileage']
```

```python
df['Basic_Warranty']=df['Basic_Warranty'].str.strip('(years/distance whichever comes first)')
df['Fuel_Tank_Capacity']=df['Fuel_Tank_Capacity'].str.replace('litres','L')
df.head()
```

```python
df['Ex-Showroom_Price'].dtype
```

```python
df['Ex-Showroom_Price'] = df['Ex-Showroom_Price'].astype(int)
```

```python
company_sales=df.groupby('Make')['Ex-Showroom_Price'].sum()
sorted_companies=company_sales.sort_values(ascending=False)
top10_companies=sorted_companies.head(10)
top10_companies
```

##**Data Vizualisation**


```python
# Set the figure size for the plot
plt.figure(figsize=(12, 6))

# Plot the data in a bar chart
top10_companies.plot(kind='bar', color='skyblue')

plt.title('Top 10 Car Companies by Sales')
plt.xlabel('Company')
plt.ylabel('Total Sales')
plt.show()
```

###Sorting

```python
plt.figure(figsize=(8, 8))
# Choose a Seaborn color palette
custom_palette = sns.color_palette('Set2')

# Plot the count of each 'Make' category with the chosen color palette
sns.countplot(y='Make', data=df, order=df['Make'].value_counts().index, palette=custom_palette)
# Set the title of the plot
plt.title('Count of Cars by Car Companies')

# Display the plot
plt.show()
```

```python
plt.figure(figsize=(8, 8))
# Choose a Seaborn color palette
custom_palette = sns.color_palette('viridis')

# Plot the boxplot with the chosen color palette
sns.boxplot(y='Make', x='Cylinders', data=df, palette=custom_palette)
```

```python
corr= df.corr()
corr
```

```python
sns.heatmap(corr,annot=True)
```

```python
plt.figure(figsize=(10, 6))

custom_palette = sns.color_palette('pastel')

ax = sns.barplot(x='Seating_Capacity', y='Make', data=df, palette=custom_palette, ci=None)

# Add annotations to the bars
for p in ax.patches:
    ax.annotate(f'{p.get_width():.0f}', (p.get_width(), p.get_y() + p.get_height() / 2), ha='left', va='center', color='black')

# Set the title of the plot
plt.title('Average Seating Capacity by Car Makers')

# Set labels for the x and y axes
plt.xlabel('Average Seating Capacity')
plt.ylabel('Car Companies')
plt.yticks(fontsize=8)

plt.show()
```

```python
fig=plt.figure(figsize=(8,6))
colors = ['#ff9999', '#66b3ff', '#99ff99', '#ffcc99']
ax=fig.subplots()
df.Fuel_Type.value_counts().plot(ax=ax,kind='pie',colors=colors)
ax.set_ylabel("")
ax.set_title("Cars count wrt to Engine fuel type")
plt.show()
```

```python

# Select only specific unique values from the 'Body_Type' column
selected_body_types = ['Hatchback', 'Sedan', 'SUV', 'MUV', 'Coupe', 'Convertible', 'Pick-up']
filtered_df = df[df['Body_Type'].isin(selected_body_types)]

# Plotting the scatter plot with seaborn
plt.figure(figsize=(12, 8))
sns.scatterplot(x='Displacement', y='Ex-Showroom_Price', hue='Body_Type', data=filtered_df, palette='viridis', s=100)
plt.title('Scatter Plot for Power vs Price with respect to each Body Type')
plt.xlabel('Power')
plt.xticks(rotation=90,fontsize=5)
plt.ylabel('Price')
plt.show()

```

```python
# Set the figure size for the plot
plt.figure(figsize=(10, 10))

# Create a pair plot with custom settings
sns.pairplot(data=df, hue='Fuel_Type', palette='Set2', markers=['o', 's', 'D', '^'], diag_kind='kde')

# Add a title to the plot
plt.suptitle("Pair Plot of Car Features")

# Display the plot
plt.show()
```
